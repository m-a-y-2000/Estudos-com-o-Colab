```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Função para obter o conteúdo do link pop-up
def get_popup_content(url):

    if url.startswith('#'):
      return None, None, None, None

    # Corrigir URL's relativas
    if not url.startswith('http'):
        url = 'https://sucupira.capes.gov.br' + url

    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    ementa = soup.find('span', id='form:ementa')
    ies = soup.find('span', id='form:ies')
    programa = soup.find('span', id='form:programa')
    nome = soup.find('span', id='form:nome')
    ementa_text = ementa.get_text(strip=True) if ementa else None
    ies_text = ies.get_text(strip=True) if ies else None
    programa_text = programa.get_text(strip=True) if programa else None
    nome_text = nome.get_text(strip=True) if nome else None
    return ementa_text, ies_text, programa_text, nome_text

# Carregar o conteúdo do arquivo HTML
with open('/content/pagina2.html', 'r', encoding='utf-8') as file:
    content = file.read()

soup = BeautifulSoup(content, 'html.parser')

# Encontrar todos os links para os pop-ups
popup_links = soup.find_all('a', href=True, onclick=True)

# Coletar informações das pop-ups
data = []
for link in popup_links:
    popup_url = link['href']
    ementa_content, ies_content, programa_content, nome_content = get_popup_content(popup_url)
    if ementa_content or ies_content or programa_content or nome_content:
        data.append({'IES': ies_content, 'Programa': programa_content, 'Disciplina': nome_content, 'Ementa': ementa_content})

# Criar um DataFrame do pandas
df = pd.DataFrame(data)

# Exibir a tabela
print(df)

# Salvar a tabela em um arquivo CSV
df.to_csv('/content/ementas.csv', index=False, encoding='utf-8')
```
