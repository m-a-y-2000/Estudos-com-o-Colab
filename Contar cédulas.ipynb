```python
def Saque():
  valor = int(input("Digite o valor a sacar: "))

  val100 = valor // 100
  resto = valor % 100

  val50 = resto // 50
  resto = resto % 50

  val10 = resto // 10
  resto = resto % 10

  val5 = resto // 5
  resto = resto % 5

  val1 = resto // 1

  print(val100, val50, val10, val5, val1)
```

```python
Saque()
```
