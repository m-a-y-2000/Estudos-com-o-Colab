```python
def NovoSalario():
  sal = float(input("Digite o valor do seu salário: "))
  porc = float(input("Digite a porcentagem do aumento: "))

  aumento = (sal*porc)/100
  novosal = sal + aumento

  print("O salário novo é: ", novosal, " e o aumento foi de: ", aumento)
```

```python
NovoSalario()
```
